# **Lecture 10 Introduction to Pandas**

## Introduction

Pandas is an open-source library that provides high-performance, easy-to-use data structures and data analysis tools for Python. It is particularly well-suited for working with tabular data, such as data stored in spreadsheets and databases. Pandas is built on top of the NumPy library and integrates well with other libraries in the Python ecosystem. For example, it helps the handling of input data stored in CSV formats and assists in converting your data into a format suitable for input into machine learning models.

If you work with Kaggle competitions, oftentimes, you will have to employ pandas to deal with the data.

````python
import numpy as np
import pandas as pd
````

## Creating a DataFrame

The fundamental data structures in pandas are created using two classes:

- DataFrame: which you can think of as a table with rows and columns having specific names.
- Series: representing a single column. A DataFrame includes one or more Series, each with its own name.

The data frame is a commonly used abstraction for data manipulation. Similar implementations are used in Spark and R.

One way to create a `Series` is to construct a `Series` object. For example:

````python
city_names = pd.Series(['Stony Brook', 'East Setauket', 'Centereach'])
population = pd.Series([12222, 12345, 23456])
````

````python
city_names
````

`DataFrame` objects can be created by passing a `dict` mapping `string` column names to their respective `Series`. If the `Series` don't match in length, missing values are filled with NA/NaN.

````python
pd.DataFrame({ 'City name': city_names, 'Population': population })
````

`DataFrames` can also be created without `Series`. For example:

````python
# Creating lists for the DataFrame
city_names = ['Stony Brook', 'East Setauket', 'Centereach']
population = [12222, 12345, 23456]

# Creating the DataFrame
pd.DataFrame({ 'City name': city_names, 'Population': population})
````

It can also take numpy arrays; however, Numpy arrays are homogeneous in type.

# 2D Poisson Implementation (optional, for those who are interested, not covered in class)

Consider a 2D Poisson's equation with Dirichlet boundary conditions:
\begin{eqnarray*}
    \Delta u(x, y) = f(x, y),& \quad& \text { in } \Omega:=(0, 1) \times (0, 1), \\
    u(x, y) = g(x, y),& \quad&  \text { on } \partial \Omega,
\end{eqnarray*}
where $\Delta$ is the Laplacian operator, $f(x) = \sin(x-y)$, and $g(x, y) = 1$.
Approximate the solution using Finite Difference Method with a grid-size $n =100$. The resulting linear system of equations $Au =b$ should be solved via Jacobi Method initialized with all 0s. The stoping criteria for Jacobi method is when $|u_{i,j}^{k+1} - u_{i,j}^{k}| < \epsilon,~ \forall i,j$. Use $\epsilon = 10e-7$. (i.e., the difference between the updated values and the old values are small).

Graph the final solution (heatmap, contour plot, or 3D plot, whichever is fine).

Recall from the lecture:
1. The unit square domain can be discretized as an $N \times N$ grid:
$$x_i = i\cdot h,~ y_j = j\cdot h,~i, j \in[0, N],$$
where the step size $h = dx = dy = \frac{1}{N}$, and
$$u_{i, j}=u\left(x_i, y_j\right), ~f_{i, j}=f\left(x_i, y_j\right) = sin(x_i - y_j).$$

2. By central difference approximation:
$$\frac{u_{i+1, j}+u_{i-1, j}-2 u_{i, j}}{h^2}+\frac{u_{i, j+1}+u_{i, j-1}-2 u_{i, j}}{h^2}=\sin \left(x_i-y_j\right)$$.

3. For all unknown $u_{i,j}, i, j \in[1, N-1]$, we have a system of $(N-2)^2$ equations with $(N-2)^2$ unknown variables, which can be written in the form of $Au = b$.

4. Solve the algebraic system numerical methods (Jacobi for this assignment).

5. The element-based Jacobi updating formula for $u_i$ is:
$$
u_i^{(k+1)}=\frac{1}{a_{i i}}\left(b_i-\sum_{j \neq i} a_{i j} u_j^{(k)}\right), \quad i=1,2, \ldots, n .
$$

Hint: The update rule for every grid point in the domain is the same so that you don't actually have to flatten $u$ and $b$ and construct the matrix $A$ although it is also okay to do so.

The contour plot of the solution looks like this:

![link text](https://raw.githubusercontent.com/wenhangao21/AMS595-Teaching/main/fdm1.png)

You can get this solution using Mathematica with the following code:
```
uval = NDSolveValue[{D[u[x, y], x, x] + D[u[x, y], y, y] ==
    Sin[(x - y)], u[x, 0] == u[x, 1] == u[0, y] == u[1, y] == 1},
  u, {x, 0, 1}, {y, 0, 1}]
ContourPlot[uval[x, y], {x, 0, 1}, {y, 0, 1}, PlotLegends -> True]
```


````python
import numpy as np
import matplotlib.pyplot as plt

def jacobi(grid, f, dx):
  # update rules
    newgrid = np.ones(shape=grid.shape)

    # apply evolution operator/ Jacobi updates
    newgrid[1:-1,1:-1] = 0.25 * (grid[1:-1,:-2] + grid[1:-1,2:] +
                                 grid[:-2,1:-1] + grid[2:,1:-1] - f[1:-1,1:-1]*dx**2)
    return newgrid

def initgrid(gridsize):
  # initialize the solution grid
    u = np.zeros((gridsize,gridsize))
    # boundary condition
    u[0,:]  = 1
    u[:,-1] = 1
    u[-1,:] = 1
    u[:,0]  = 1
    return u

n = 100  # grid-size
u = initgrid(n)   # initial grid, n by n np array
L = 1.0    # square bdy length
dx = L/(n-1)   # FDM step size
x = np.linspace(0, L, n)
X, Y = np.meshgrid(x, x)  # X, Y grid
f = np.sin(X-Y)    # RHS values, n by n np array
epsilon = 1e-7
iter = 0

while True:
    if iter % 500 ==0:
        u_old = u
    u = jacobi(u, f, dx)
    # check this condition every 500 iterations
    if iter % 500 == 0:
        diff = np.absolute(u_old - u).max()
        if diff < epsilon:
            break
    iter = iter + 1

plt.figure(figsize=(7,7))
plt.contourf(X, Y, u)
plt.gca().set_aspect('equal')
plt.colorbar()
````
